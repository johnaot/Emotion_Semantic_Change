```python
# look at nearest neighbours
t1, t2 = 9, -1
k = 100
k_nns = 3
pos = NOUN

time_range = list(range(1800, 2000, 10))
def inspect_nns(changes, pos, pos_data, vectors, t1, t2, time_range, k_nns):
    for w in sorted(list(changes.keys()), key=lambda w: changes[w]):
        first_nns = get_closest_neighbours(w, pos, pos_data, vectors, t1, k_nns)
        last_nns = get_closest_neighbours(w, pos, pos_data, vectors, t2, k_nns)
        print(w, time_range[t1], first_nns, time_range[t2], last_nns)

print('English emotion')
inspect_nns(measures_emotion_eng[k], pos, pos_data, data, t1, t2, time_range, k_nns)
print('--------------------')
print('French emotion')
inspect_nns(measures_emotion_fra[k], pos, pos_data_fr, data_fr, t1, t2, time_range, k_nns)
```

```python
# plot correlations matrices
from scipy.stats.stats import pearsonr, spearmanr

def plot_matrix(axes, measures_lst_lst, labels, name, use_bar=True):
    corrs, pvals = [], []
    for lst1 in measures_lst_lst:
        corrs_lst, pvals_lst = [], []
        for lst2 in measures_lst_lst:
            corr, pval = pearsonr(lst1, lst2)
            corrs_lst.append(corr) 
            pvals_lst.append(pval)
        corrs.append(corrs_lst)
        pvals.append(pvals_lst)
    corrs = np.array(corrs).round(decimals=3)
    pvals = np.array(pvals).round(decimals=6)
    
    for i, ax in enumerate(axes):
        matrix = [corrs,pvals][i]
        print(matrix)
        im = ax.imshow(matrix, vmin=0, vmax=1.)
        if use_bar:
            cbar = ax.figure.colorbar(im, ax=ax)
            cbar.ax.tick_params(labelsize=20) 
        ax.set_xticks(np.arange(len(labels)))
        ax.set_yticks(np.arange(len(labels)))
        ax.set_xticklabels(labels)
        ax.set_yticklabels(labels)
        ax.xaxis.set_tick_params(labelsize=28)
        ax.yaxis.set_tick_params(labelsize=28)
        ax.grid(False)
        for i in range(len(labels)):
            for j in range(len(labels)):
                text = ax.text(j, i, matrix[i, j], ha="center", va="center", color="w", fontsize=20)

def figure_matrix_labels(axes, labels):
    axes[0].annotate(labels[0],
            xy=(0.5, 0.95), xytext=(0.5, 0.95),
            xycoords=('axes fraction', 'figure fraction'),
            textcoords='offset points',
            fontsize=35, weight='bold', ha='center', va='center')
    axes[1].annotate(labels[1],
            xy=(0.5, 0.95), xytext=(0.5, 0.95),
            xycoords=('axes fraction', 'figure fraction'),
            textcoords='offset points',
            fontsize=35, weight='bold', ha='center', va='center')
    return axes

labels = [str(k) for k in ks]

fig, axes = plt.subplots(1, 2, sharex=False, sharey=False, figsize=(18, 8))
plot_matrix([axes[0]], [[measures_emotion_eng[k][w] for w in measures_emotion_eng[k]] for k in ks], labels, ['English', 'emotion', 'words'], use_bar=True)
plot_matrix([axes[1]], [[measures_emotion_fra[k][w] for w in measures_emotion_fra[k]] for k in ks], labels, ['French', 'emotion', 'words'], use_bar=True)
axes = figure_matrix_labels(axes, ['English', 'French'])
format_name = 'svg'
plt.tight_layout()
plt.savefig('images/corrs_%s.%s' % ('emotion',format_name), format=format_name, bbox_inches='tight')

```
